```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pylab
import random
```

```python
# Task 1, n = 10
sample1 = []
heads = 0

for y in range(1000):
    for x in range(10):
        heads = heads + random.randint(0,1)
    sample1.append(heads)
    heads = 0
    
plt.hist(sample1, bins=9, edgecolor='black')
plt.xlabel('# of Heads')
plt.ylabel('Trials')
plt.show()
```
