```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.stats import norm

# Task 1, n = 10
sample1 = []
heads = 0

for y in range(1000):
    for x in range(10):
        heads = heads + random.randint(0,1)
    sample1.append(heads)
    heads = 0

# Plot histogram
plt.hist(sample1, bins=20, edgecolor='black', density=True, alpha=0.75, label='Histogram')

# Calculate mean and standard deviation
mean = np.mean(sample1)
std_dev = np.std(sample1)

# Create a normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)

# Plot normal curve
plt.plot(x, p, 'k', linewidth=2, label='Normal Curve')

plt.xlabel('# of Heads')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
